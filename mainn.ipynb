```python
import pandas as pd  

sentim = pd.read_csv('sentiments.txt', sep = ';', names = ['sentence','emotion'])
print(sentim.head())
print(sentim.info())
```

```python
print(sentim['emotion'].value_counts())

#sentim.loc[sentim['emotion'] == 'joy', 'emotion'] = 1

#print(sentim['emotion'].value_counts())

sentim.replace({'joy': 1, 'love': 1, 'surprise': 1, 'sadness': 0, 'anger': 0, 'fear': 0}, inplace = True)
print(sentim['emotion'].value_counts())
```

```python
import re  
import nltk   

nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords  
from nltk.stem import WordNetLemmatizer
```

```python
lemm = WordNetLemmatizer()

def transformtxt(txt):
    stop = set(stopwords.words('english'))
    newtxt = []

    for sentence in txt:
        sentence = re.sub('[^a-zA-z]', ' ', sentence)
        sentence = sentence.lower()
        sentence = sentence.split()

        newwords = []

        for word in sentence:
            if word not in stop:
                newword = lemm.lemmatize(word)
                newwords.append(newword)

        newsentence = ' '.join(newwords)
        newtxt.append(newsentence)
    return newtxt


```

```python
x = sentim['sentence']
y = sentim['emotion']
```

```python
newx = transformtxt(x)

print(newx[0:10])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countv = CountVectorizer(ngram_range = (1,2))

newwx = countv.fit_transform(newx)

print(newwx.shape)
print(newwx[1])
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(newwx, y, test_size = 0.2, random_state = 40)
```

```python
from sklearn.ensemble import RandomForestClassifier
ranforest = RandomForestClassifier()
ranforest.fit(x_train, y_train)

y_pred = ranforest.predict(x_test)

print(y_pred)
```
